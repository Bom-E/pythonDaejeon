```python
load_dotenv()

url = 'http://apis.data.go.kr/6300000/openapi2022/safeCCTV/getsafeCCTV'
params = {
    'serviceKey' : os.getenv('my_service_key')
    , 'numOfRows' : '2985'
    , 'pageNo' : '1'
}

# .env 적용 드디어 해결! 이거 해결 하지 전에는 안 자려고 했는데 이제 잔다!

response = requests.get(url, params=params)

result_data = response.json()

# print(result_data)

items = result_data['response']['body']['items']

# for item in items:
    # print(f"관리번호 : {item['manageNo']}, 주소 : {item['lnmAdres']}, x좌표 : {item['crdntX']}, y좌표 : {item['crdntY']}")

# 글 길이가 제멋대로라서 간단한 replace로 불가능.(가능 할 것 같은데, 우악스러워질듯) 
# 데이터의 구성을 보니
# 1. . 을 기준으로 no를 부여
# 2. 처음 나오는 ()로 cctv 관할구와 숫자범위를 부여
# 3. 지도상 주소명

# 따라서 )를 기준으로 나누기.

# 데이터의 구조 : 'manageNo': '091.(서대전 88-2)동성이용원(서대전초교 뒷길)(회전)'
# 구현 이유 : 사용자가 이해할 만한 cctv의 자연어 명칭 뽑아내기

def adres_name(manage_no):
    # ')' 기준 분할 (cctv 명칭 데이터)
    adres_name = manage_no.split(')', 1)
    return adres_name[1] if len(adres_name) > 1 else manage_no.split('.', 1)[1]

def transform_item(item):
    return{
        'manageNo' : item['manageNo'][0:3]
        , 'address_name' : adres_name(item['manageNo'])
        , 'address' : item['lnmAdres']
        , 'x' : item['crdntX']
        , 'y' : item['crdntY']
    }

cctv_list = list(map(transform_item, items))

# 리스트를 데이터프레임으로 바꾸기
df_cctv = pd.DataFrame(cctv_list)

df_cctv.set_index("manageNo", inplace=True)

df_cctv
```

```python
df_dict1 = pd.read_csv('data/대전광역시_유성구_여성안심지킴이집_20240214.csv', encoding="CP949")
df_dict2 = pd.read_csv('data/대전광역시_중구_여성안심지킴이집_20231130.csv', encoding="CP949")
df_dict3 = pd.read_csv('data/대전광역시_서구_여성안심지킴이집_20231109.csv', encoding="CP949")
df_dict4 = pd.read_csv('data/대전광역시_동구_여성안심지킴이집_20230721.csv', encoding="CP949")
df_dict5 = pd.read_csv('data/대전광역시_대덕구_여성안심지킴이집_20230419.csv', encoding="CP949")

df_house = [df_dict1, df_dict2, df_dict3, df_dict4, df_dict5]

for i in range(len(df_house)):
    df_house[i].columns = df_house[i].columns.str.strip()
    df_house[i] = df_house[i][df_house[i]['운영여부'] != 'N']

df_result = pd.concat(df_house, ignore_index=True)

# 지정연도, 운영여부 삭제(운영여부는 어차피 전부 Y임)
df_result = df_result.drop(columns=['지정연도', '운영여부'])

df_result

# 데이터 결측치 제거(원래 NaN 데이터를 날리려 했으나, 데이터 조회 후 결정.)
# df_result[df_result.isna().any(axis=1)]

# NaN 데이터가 있는 행만 조회
df_result[df_result.isna().any(axis=1)]

# 폐업신고 한 가게 행 삭제
drop_data = [22, 61, 74]
df_result.drop(drop_data, axis=0, inplace=True)

# 전화번호 NaN 수정
# iat으로 하니까 기존 데이터의 인덱스에 데이터가 덧씌워진다...
# 레이어 기준이 아니라 위치 기준이라서 그럴 가능성이 큰 듯. 게다가 행 삭제 하고 인덱스 재지정도 안 해줬으니...
# 폐업 파악 후 인덱스 재지정 해주자. (그러고 iat으로 잘못 지정한 전화번호 값 2개도 수정해야함...)
df_result.at[31, '여성안심지킴이집전화번호'] = '1577-8007'
df_result.at[32, '여성안심지킴이집전화번호'] = '02-6916-1500'
df_result.at[33, '여성안심지킴이집전화번호'] = '070-7740-0008'
df_result.at[34, '여성안심지킴이집전화번호'] = '전화번호 없음'
df_result.at[40, '여성안심지킴이집전화번호'] = '1577-8007'
df_result.at[42, '여성안심지킴이집전화번호'] = '1577-8007'
df_result.at[48, '여성안심지킴이집전화번호'] = '042-522-7433'
df_result.at[51, '여성안심지킴이집전화번호'] = '1577-8007'
df_result.at[52, '여성안심지킴이집전화번호'] = '1577-8007'
df_result.at[53, '여성안심지킴이집전화번호'] = '1577-8007'
df_result.at[71, '여성안심지킴이집전화번호'] = '1577-8007'
df_result.at[75, '여성안심지킴이집전화번호'] = '080-999-5425'

# 카카오주소 api를 통해 지번주소가 NaN 제거


# 주소 NaN 수정
# 아름답지 않아... 카카오주소api를 이용해서 자동화 하면 될 것 같은데... 일단 생각해보자
# 1. 카카오주소 api를  requests로 불러오기
# 2. if문으로 소재지지번주소가 NaN 값인 행들을 조회
# 3. 소재지도로명주소를 잘라서 검색 후 지번주소를 찾아서 넣기.

# 좀 더 적합한 api를 찾음 (주소기반산업지원서비스)

load_dotenv()

def get_jibun(road_address):
    url = "https://business.juso.go.kr/addrlink/addrLinkApi.do"
    params = {
        "confmKey": os.getenv('address_key'),
        "currentPage": "1",
        "countPerPage": "10",
        "keyword": road_address,
        "resultType": "json"
    }

    response = requests.get(url, params=params)
    data = response.json()

    if data['results']['common']['totalCount'] != '0':
        # null 처리
        return data['results']['juso'][0]['jibunAddr']
    else:
        return "주소를 찾을 수 없습니다."

# 데이터 확인
# road_address = "울산광역시 울주군 삼남읍 도호 1길 39-11"
# jibun_address = get_jibun(road_address)
# print(f"도로명주소: {road_address}")
# print(f"지번주소: {jibun_address}")

# 기존 데이터프레임에 소재지지번주소 데이터 추가

copy_df = df_result.copy()

mask = copy_df['소재지지번주소'].isna()

copy_df.loc[mask, '임시_소재지지번주소'] = copy_df.loc[mask, '소재지도로명주소'].apply(get_jibun)

df_result.loc[mask, '소재지지번주소'] = copy_df.loc[mask, '임시_소재지지번주소']

# print(df_result['소재지지번주소'])

del copy_df

# 전화번호가 없는 지점(결측치 처리 후 마지막에 할 것.)
# df_result['여성안심지킴이집전화번호'].fillna('전화번호 없음', inplace=True)

# 인덱스 재지정(이것도 결측치 처리 후 마지막에 할 것.)
# df_result.reset_index(drop=True, inplace=True)

df_result[df_result.isna().any(axis=1)]

df_result[df_result['소재지지번주소'] == '주소를 찾을 수 없습니다.']

# 1. 주소를 찾을 수 없습니다. 주소 찾아서 넣기, 안 된 이유가 뭔지 분석하기
# CU월평사랑점 : 도로명 주소에 건물 번호와 건물 명이 함께 붙어있음(정작 도로명 주소를 검색하면 건물 명은 안 뜸)
# GS월평타운점 : 도로명 주소에 건물 번호와 . + 층명 함께 붙어있음(정작 도로명 주소를 검색하면 번호는 안 뜸)
# 아마트24 월평중앙점 : 위와 이하동문

# 데이터 재가공
df_result.at[81, '소재지도로명주소'] = '대전 서구 월평북로 27 (월평동) 1층 101호'
df_result.at[81, '소재지지번주소'] = '대전광역시 서구 월평동 218-6'
df_result.at[82, '소재지도로명주소'] = '대전광역시 서구 월평서로6번길 26'
df_result.at[82, '소재지지번주소'] = '대전광역시 서구 월평동 1069'
df_result.at[83, '소재지도로명주소'] = '대전광역시 서구 계룡로231번길 9'
df_result.at[83, '소재지지번주소'] = '대전광역시 서구 월평동 1407'

# 데이터 확인
df_result[df_result['점포명'] == 'CU 월평사랑점']

df_result[df_result.isna().any(axis=1)]
```
